```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Basic_tasks_DA\Third_task\iris.csv")
```

Bar Plot


```python
plt.Figure(figsize=(8,5))
df['petal_length'].value_counts().plot(kind='bar',color='purple')
plt.title("iris dataset")
plt.xlabel("lala")
plt.ylabel("count")
plt.tight_layout()
plt.savefig("bar_chart.png")
plt.show()
```

Line chart

```python
plt.Figure(figsize=(10,7))
plt.plot(df['petal_length'], df['petal_width'],color='magenta')
plt.title("Line plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.tight_layout()
plt.savefig("line_chart.png")
plt.show()
```

Scatter Plot

```python
plt.Figure(figsize=(10,7))
plt.scatter(df['sepal_length'], df['sepal_width'], color='red')
plt.title("Scatter Plot")
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.tight_layout()
plt.savefig("scatter_plot.png")
plt.show()
```
